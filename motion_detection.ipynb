```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def adjusted_luminosity(x):

    b = np.multiply(x[:,:,0], 0.11)
    g = np.multiply(x[:,:,1], 0.59)
    r = np.multiply(x[:,:,2], 0.3)

    gray = np.add(np.add(b, g), r)

    return gray
```

```python
gif = cv.VideoCapture("images/armor.gif")
imlist = []

flag = True

while flag:
    
    ret, frame = gif.read()
    imlist.append(frame)

    if not ret:
        break

imlist = np.array(imlist)
imlist.shape
```

```python
print(type(imlist[0]))
print(imlist[0].shape)
```

```python
graylist = []
for i in range(len(imlist)):
    if i == 250:
        break
    graylist.append(adjusted_luminosity(imlist[i]))

plt.imshow(graylist[0], cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
f = plt.figure(figsize=(20,8))
ax1 = f.add_subplot(131)
ax1.imshow(graylist[0], cmap='gray', vmin=0, vmax=255)
ax1.set_title("Frame 0")

ax2 = f.add_subplot(132)
ax2.imshow(graylist[10], cmap='gray', vmin=0, vmax=255)
ax2.set_title("Frame 10")

ax3 = f.add_subplot(133)
ax3.imshow(graylist[20], cmap='gray', vmin=0, vmax=255)
ax3.set_title("Frame 20")

plt.show()
```
