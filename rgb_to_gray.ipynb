## Average Grayscale

grayscale = (Red + Green + Blue) / 3

```python
def grayscale(x):
    
    b = x[:,:,0]
    g = x[:,:,1]
    r = x[:,:,2]

    gray = np.divide(np.add(np.add(b, g), r), 3)

    return gray

plt.figure(figsize=(11,6))
plt.imshow(grayscale(image), cmap='gray', vmin=0, vmax=255)
plt.show()
```

## Weighted average or adjusted luminosity method

grayscale = ( (0.3 * Red) + (0.59 * Green) + (0.11 * Blue) )

```python
def adjusted_luminosity(x):

    b = np.multiply(x[:,:,0], 0.11)
    g = np.multiply(x[:,:,1], 0.59)
    r = np.multiply(x[:,:,2], 0.3)

    gray = np.add(np.add(b, g), r)

    return gray


plt.figure(figsize=(11,6))
plt.imshow(adjusted_luminosity(image), cmap='gray', vmin=0, vmax=255)
plt.show()
```
