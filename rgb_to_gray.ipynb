```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("leaf.jpg")
```

```python
plt.figure(figsize=(11,6))
plt.imshow(image)
plt.show()
```

```python
type(image)
```

```python
b = image[:,:,0]
g = image[:,:,1]
r = image[:,:,2]
```

```python
f = plt.figure(figsize=(20,8))
ax1 = f.add_subplot(131)
ax1.imshow(b)
ax1.set_title("Blue Channel")

ax2 = f.add_subplot(132)
ax2.imshow(g)
ax2.set_title("Green Channel")

ax3 = f.add_subplot(133)
ax3.imshow(r)
ax3.set_title("Red Channel")

plt.show()
```

```python
def grayscale(x):
    
    b = x[:,:,0]
    g = x[:,:,1]
    r = x[:,:,2]

    gray = np.divide(np.add(np.add(b, g), r), 3)

    return gray

plt.figure(figsize=(11,6))
plt.imshow(grayscale(image))
plt.show()
```
